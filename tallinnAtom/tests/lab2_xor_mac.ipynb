```python
base_key_str = "test"
secret_key = 0
for b in base_key_str.encode():
    secret_key ^= b

def xor_mac(message, key):
    acc = 0
    for b in message:
        acc ^= b
    acc ^= key
    return acc

secret_key

```

```python
message = b"generate"
tag = xor_mac(message, secret_key)

message, secret_key, tag


```

```python
found_key = None
for guess in range(256):
    if xor_mac(message, guess) == tag:
        found_key = guess
        break

secret_key, found_key, (secret_key == found_key)
```

```python
malicious_message = b"FORMAT_FLASH"
malicious_tag = xor_mac(malicious_message, found_key)

malicious_message, malicious_tag

```

```python
def verify(message, tag, key):
    return xor_mac(message, key) == tag

verify(malicious_message, malicious_tag, secret_key)

```
