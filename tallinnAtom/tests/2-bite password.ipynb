```python
import hmac, hashlib

def hmac_sha256(message, key_int):
    key_bytes = key_int.to_bytes(2, "big")
    return hmac.new(key_bytes, message, hashlib.sha256).digest()

message = b"hack"
tag = bytes.fromhex("ad8c2e78c881a1785d774e0cf4696b1b6e6857120eebdf2d2596404883bb9c4b")

message, tag.hex()
```

```python
found_key = None

for guess in range(65536):
    if hmac_sha256(message, guess) == tag:
        found_key = guess
        break

found_key


```
